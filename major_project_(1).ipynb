```python
import matplotlib.pyplot as plt
plt.scatter(df["Weightgainedgrams"],df["CaloriesConsumed"])
plt.title("Weightgainedgrams v/s CaloriesConsumed")
plt.xlabel("Weightgainedgrams")
plt.ylabel("CaloriesConsumed")
```

```python
x = df.iloc[:,0:1].values
x
```

```python
y = df.iloc[:,0].values
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
model.predict([[108]])
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
m*108 +c
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,c="red")
plt.title("Weightgainedgrams v/s CaloriesConsumed")
plt.xlabel("Weightgainedgrams")
plt.ylabel("CaloriesConsumed")
```
