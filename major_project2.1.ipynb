```python
#EDGE DETECTION -  3 main types
#1.SOBEL #2.LAPLACIAN #3.CANNY(most accurate one)
import cv2
img = cv2.imread('123.png',0) #0 converts image into grayscale
sobel_x = cv2.Sobel(img,cv2.CV_8U,dx = 1,dy = 0,ksize = -1)
sobel_y = cv2.Sobel(img,cv2.CV_8U,dx = 0,dy = 1,ksize = -1)
sobel = cv2.bitwise_or(sobel_x,sobel_y) 
cv2.imshow('Sobel X',sobel_x)
cv2.imshow('Sobel Y',sobel_y)
cv2.imshow('SOBEL',sobel)
laplacian = cv2.Laplacian(img,cv2.CV_8U)
cv2.imshow('Laplacian',laplacian)
canny = cv2.Canny(img,90,200)
cv2.imshow('Canny Image',canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#SCALING/RESIZING of an IMAGE
import cv2
img = cv2.imread('C:\Users\ASUS\Desktop\python\rinex_opencv\bird.jpg') #read the image
cv2.imshow('ORIGIANL IMAGE',img) #display the image
img1 = cv2.resize(img,None,fx = 0.75,fy = 0.75)
cv2.imshow('Scaled Down Image',img1)
img2 = cv2.resize(img,None,fx = 1.5,fy = 1.5)
cv2.imshow('Scaled Up Image',img2)
img3 = cv2.resize(img,(1000,400))
cv2.imshow('Custom dimensions',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#.FACE DETECTION
import cv2
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('C:\Users\ASUS\Desktop\python\rinex_opencv\man.jpg') #reading the image
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray,scaleFactor = 1.1,minNeighbors = 9)
for x,y,w,h in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
cv2.imshow('Face Detection',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#FACE and EYE DETECTION
import cv2
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eyes_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")
img = cv2.imread("C:\Users\ASUS\Desktop\python\rinex_opencv\man.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray,scaleFactor = 1.1,minNeighbors =9)
eyes = eyes_cascade.detectMultiScale(gray,scaleFactor = 1.1,minNeighbors = 18)
for x,y,w,h in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
for x,y,w,h in eyes:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),5) 
cv2.imshow("face detection",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
